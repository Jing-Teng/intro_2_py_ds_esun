## 自訂函數 sign_fun() 並且作圖（自由取一段 x 區間）

```python
import numpy as np
import matplotlib.pyplot as plt

def sign_fun(x):
    if x < 0:
        return -1
    elif x == 0:
        return 0
    else:
        return 1

x_arr_1 = np.linspace(-5, -0.1, 50)
x_arr_2 = 0
x_arr_3 = np.linspace(0.1, 5, 50)
y_arr_1 = map(sign_fun, x_arr_1)
y_arr_2 = 0
y_arr_3 = map(sign_fun, x_arr_3)
plt.plot(x_arr_1, y_arr_1, c = "b")
plt.scatter(x_arr_2, y_arr_2, c = "b")
plt.scatter([0, 0], [-1, 1], facecolors = 'none', edgecolors = 'b')
plt.plot(x_arr_3, y_arr_3, c = "b")
plt.xticks([0], ["0"]), plt.yticks([-1, 0, 1], ["-1", "0", "1"])
plt.title("$y = sign(x)$")
plt.show()
```

## 自訂函數 sigmoid_fun() 並且作圖（自由取一段 x 區間）

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_fun(x):
    return 1 / (1 + np.exp(-x))

x_arr = np.linspace(-5, 5, 100)
y_arr = map(sigmoid_fun, x_arr)
plt.plot(x_arr, y_arr)
plt.xticks([0], ["0"]), plt.yticks([0, 0.5, 1], ["0", "0.5", "1"])
plt.axvline([0], linestyle = "--", c = "k", alpha = 0.5)
plt.axhline([0.5], linestyle = "--", c = "k", alpha = 0.5)
plt.title("$y = S(x)$")
plt.show()
```

## 承前題，將前兩個函數畫在同一個函數圖（Optional）

```python
x_arr_1 = np.linspace(-5, -0.1, 50)
x_arr_2 = 0
x_arr_3 = np.linspace(0.1, 5, 50)
y_arr_1 = map(sign_fun, x_arr_1)
y_arr_2 = 0
y_arr_3 = map(sign_fun, x_arr_3)
plt.plot(x_arr_1, y_arr_1, c = "b")
plt.scatter(x_arr_2, y_arr_2, c = "b")
plt.scatter([0, 0], [-1, 1], facecolors = 'none', edgecolors = 'b')
plt.plot(x_arr_3, y_arr_3, c = "b")
x_arr = np.linspace(-5, 5, 100)
y_arr = map(sigmoid_fun, x_arr)
plt.plot(x_arr, y_arr, c = "r", linestyle = "-")
plt.xticks([0], ["0"]), plt.yticks([-1, 0, 0.5, 1], ["-1", "0", "0.5", "1"])
plt.title("Sign and Sigmoid")
plt.show()
```
